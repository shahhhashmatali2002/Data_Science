```python
#How to handle missing data
```

```python
df = pd.read_csv('new1.csv',skiprows=1)
```

```python
df.head()
```

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
    'eps': 0,
    'revenue': 0,
    'price': 0,
    'people': 'no value'
})
new_df
```

```python
# empty cell value fill from back vlaue column
new_df = df.fillna(method="ffill")
new_df
```

```python
# empty cell value fill from next vlaue column
new_df = df.fillna(method="bfill")
new_df
```

```python
# empty cell value fill from next vlaue column
new_df = df.fillna(method="bfill", limit=1)
new_df
```

```python
# column drop if any cell is empty 
new_df = df.dropna()
new_df
```

```python
# if column has valid value according to thresh then drop otherwise don't drop 
new_df = df.dropna(thresh=4)
new_df
```

```python
new_df = df.dropna(thresh=2)
new_df
```

```python
#Concatenate Dataframe
pak_weather = pd.DataFrame({
    'city': ["Sindh","Balochistan","Punjab","Peshawur"],
    'temperature': [47,30,40,50],
    "humadity": [68,54,45,65]
})
pak_weather
```

```python
china_weather = pd.DataFrame({
    'city': ["Beijing","Chongqing","Shanghai","Tianjin"],
    'temperature': [47,30,40,50],
    "humadity": [68,54,45,65]
})
china_weather
```

```python
df_concat = pd.concat([pak_weather, china_weather])
df_concat
```

```python
df_concat = pd.concat([pak_weather, china_weather], ignore_index=True)
df_concat
```

```python
df_concat = pd.concat([pak_weather, china_weather], keys=['Pak',"China"])
df_concat
```
