```python
def convert_people_cell(cell):
    if cell == "n.a.":
        return 'sam walton'
        print('ture')
    return cell

df = pd.read_excel('stock.xlsx','new',converters={
    "col4": convert_people_cell,
})
```

```python
#How to handle missing data
```

```python
df = pd.read_csv('new1.csv',skiprows=1)
```

```python
df.head()
```

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
    'eps': 0,
    'revenue': 0,
    'price': 0,
    'people': 'no value'
})
new_df
```

```python
# empty cell value fill from back vlaue column
new_df = df.fillna(method="ffill")
new_df
```

```python
# empty cell value fill from next vlaue column
new_df = df.fillna(method="bfill")
new_df
```
