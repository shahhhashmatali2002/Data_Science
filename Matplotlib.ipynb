```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [41,31,32,38,37,39,36]
avg_t = [20,21,22,28,27,29,26]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")
plt.legend(loc="lower right",shadow=True,fontsize="small")
```

```python
# Bar Chat 
company = ['Google',"AMZN","MSFT","FB"]
revenue = [90,136,89,27]
profit = [40,2,34,12]
```

```python
ypos = np.arange(len(company))
ypos
xpos = np.arange(len(company))
```

```python
plt.xticks(ypos,company)
plt.ylabel('revenue(bin)')
plt.title('US Tech Stocks')
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="Profit")
plt.legend()
```

```python
plt.yticks(ypos,company)
plt.ylabel('revenue(bin)')

plt.barh(xpos-0.2,revenue,label="Revenue")
plt.barh(xpos+0.2,profit,label="Profit")
plt.legend()
```

```python
# Historgram 
blood_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women = [63,98,89,45,56,90,112,120,80,70,80,100]

plt.xlabel('sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['green','orange'],label=['men','women'])
plt.legend()
```
